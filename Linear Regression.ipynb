```python
x=np.array([1,2,3,4])
b=[2]
m=2
y=m*x+b
y
```

```python
w=np.zeros(1)
b1=0
iterations=10
def update_weights(x,w,b1):
    y_pred=x.dot(w)+b
    dw=-(2*(x.T).dot(y-y_pred))/4
    db=-2*np.sum(y-y_pred)/4
    w=w-0.1*dw
    b1=b1-0.1*db
    return w,b
for i in range(iterations):
    w,b1=update_weights(np.array([1,2,3,4]).reshape(-1,1),w,b1)
```
